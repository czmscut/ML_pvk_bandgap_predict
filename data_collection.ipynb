```python
ele_pro = pd.read_csv('element_properties.csv')

ele = ele_pro['符号']
electronegativity = ele_pro['电负性']
weight = ele_pro['原子量']
affinity = ele_pro['电子亲合能']
fir_ionization_energy = ele_pro['第一电离能']


A_elec, B_elec, X_elec = [], [], []
A_weight, B_weight, X_weight = [], [], []
A_affinity, B_affinity, X_affinity = [], [], []
A_fir_ion_energy, B_fir_ion_energy, X_fir_ion_energy = [], [], []


for i in df["A_site_ele"]:
    ind = list(ele).index(i)
    A_elec.append(electronegativity[ind])
    A_weight.append(weight[ind])
    A_affinity.append(affinity[ind])
    A_fir_ion_energy.append(fir_ionization_energy[ind])

for i in df["B_site_ele"]:
    ind = list(ele).index(i)
    B_elec.append(electronegativity[ind])
    
    B_weight.append(weight[ind])
    B_affinity.append(affinity[ind])
    B_fir_ion_energy.append(fir_ionization_energy[ind])

for i in df["X_site_ele"]:
    ind = list(ele).index(i)
    X_elec.append(electronegativity[ind])
    X_weight.append(weight[ind])
    X_affinity.append(affinity[ind])
    X_fir_ion_energy.append(fir_ionization_energy[ind])
```

```python
ion_radius = pd.read_csv('ions-radius.csv')
ion = ion_radius['ions']
radius = ion_radius['radius/pm']
A_radius, B_radius, X_radius = [], [], []

for i in df['A_site_ele']+df['A_site_oxid']:
    ind = list(ion).index(i)
    A_radius.append(radius[ind])

for i in df['B_site_ele']+df['B_site_oxid']:
    ind = list(ion).index(i)
    B_radius.append(radius[ind])

for i in df['X_site_ele']+df['X_site_oxid']:
    ind = list(ion).index(i)
    X_radius.append(radius[ind])

df['A_site_radius']=A_radius
df['B_site_radius']=B_radius
df['X_site_radius']=X_radius
```

```python
df['A_site_electronegativity']=A_elec
df['B_site_electronegativity']=B_elec
df['X_site_electronegativity']=X_elec

df['A_site_weight']=A_weight
df['B_site_weight']=B_weight
df['X_site_weight']=X_weight

df['A_site_affinity']=A_affinity
df['B_site_affinity']=B_affinity
df['X_site_affinity']=X_affinity

df['A_site_fir_ionization_energy']=A_fir_ion_energy
df['B_site_fir_ionization_energy']=B_fir_ion_energy
df['X_site_fir_ionization_energy']=X_fir_ion_energy
```

```python
import numpy as np

tor_fac = []
oct_fac = []
u_t_n = []

for i in range(0,df.shape[0]):
    t = (df['A_site_radius'][i]+df['X_site_radius'][i])/(np.sqrt(2)*(df['X_site_radius'][i]+df['B_site_radius'][i]))
    tor_fac.append(t)
    u = df['B_site_radius'][i]/df['X_site_radius'][i]
    oct_fac.append(u)
    V_A = (4*np.pi*(df['A_site_radius'][i]**3))/(3*1000000)
    V_B = (4*np.pi*(df['B_site_radius'][i]**3))/(3*1000000)
    V_X = (4*np.pi*(df['X_site_radius'][i]**3))/(3*1000000)
    n = (V_A+V_B+3*V_X)/df['volume'][i]
    utn = (t+u)**n
    u_t_n.append(utn)

df['tolerance_factor']=tor_fac
df['octahedral_factor']=oct_fac
df['(u+t)n']=u_t_n
```
